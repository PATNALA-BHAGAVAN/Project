## SPLITTING THE DATA INTO TRAIN AND TEST DATA

```python
from sklearn.model_selection import train_test_split
```

```python
d = pd.read_csv("mobile_price_range_data.csv")
```

```python
X = d.iloc[:,:-1]
Y = d.iloc[:,-1]
```

```python
# splitting the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.50,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## A) LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
print('Training score', model.score(X_train, Y_train))
print('Testing score', model.score(X_test, Y_test))
```

```python
pred1 =model.predict(X_test)
pred1
```

## B) KNN CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 11)
knn.fit(X_train,Y_train)
```

```python
print('Training score', knn.score(X_train, Y_train))
print('Testing score', knn.score(X_test, Y_test))
```

```python
pred2 = knn.predict(X_test)
pred2
```

## C) SVM CLASSIFIER

```python
classes = 4
X , t = make_classification(100,5,n_classes = classes, random_state = 40, n_informative = 2, n_clusters_per_class = 1)
svm_model = svm.SVC(kernel = "linear", random_state = 0 ,C = 0.50)
svm_model.fit(X_train,Y_train)
```

```python
pred3 = svm_model.predict(X_test)
pred4 = svm_model.predict(X_train)
print(pred3)
print(pred4)
```
